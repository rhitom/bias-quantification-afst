```python
def corrTable(df, title):
    corrTable = df.corr()
    plt.figure(figsize=(10,10))
    g = sns.heatmap(corrTable, cmap='coolwarm', annot=True, )
    plt.savefig(title)
    plt.show()
    g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right')
    g.set_yticklabels(g.get_yticklabels(), rotation=45, horizontalalignment='left')
```
